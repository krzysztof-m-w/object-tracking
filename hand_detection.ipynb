```python
from utils import *
import matplotlib.pyplot as plt
```

```python
def detect_foreground(video, path):
    width = int(video.get(3))
    height = int(video.get(4))

    fps = video.get(cv2.CAP_PROP_FPS)

    background_writer = cv2.VideoWriter(
        f"foreground_detection/{path}.mp4",
        cv2.VideoWriter_fourcc(*"mp4v"),
        fps,
        (width, height),
        0,
    )

    foreground_mog2 = cv2.createBackgroundSubtractorMOG2()


    video.set(cv2.CAP_PROP_POS_FRAMES, 0)
    while video.isOpened():
        ret, frame = video.read()

        if ret:
            background_writer.write(foreground_mog2.apply(frame))
        else:
            break

    background_writer.release()
```

```python
def get_average_brightness(path):
    foreground_video = video_from_path(f'foreground_detection\{path}.mp4')

    average_foregrounds = list()

    foreground_video.set(cv2.CAP_PROP_POS_FRAMES, 0)
    while foreground_video.isOpened():
        ret, frame = foreground_video.read()

        if ret:
            average_foregrounds.append(np.average(frame))
        else: 
            break

    return average_foregrounds
```

```python
def cut_video(video, average_foregrounds, path, threshold):
    width = int(video.get(3))
    height = int(video.get(4))

    fps = video.get(cv2.CAP_PROP_FPS)

    cut_writer = cv2.VideoWriter(
        f"cut_videos/{path}.mp4",
        cv2.VideoWriter_fourcc(*"mp4v"),
        fps,
        (width, height),
        1,
    )

    i = 0
    used_frames = np.zeros(len(average_foregrounds), np.bool_)

    video.set(cv2.CAP_PROP_POS_FRAMES, 0)
    while video.isOpened():
        ret, frame = video.read()
        if ret:
            if average_foregrounds[i] < threshold:
                cut_writer.write(frame)
                used_frames[i] = True
            i += 1
        else:
            break

    cut_writer.release()
    return used_frames
```

```python
def save_used_frames(used_frames, path):
    np.save(f'used_frames/{path}.npy', used_frames)
```

### Easy1

```python
chess = video_from_path('recordings/Easy1.mp4')
detect_foreground(chess, 'Easy1')
```

```python
average_foregrounds = get_average_brightness('Easy1')
plt.plot(average_foregrounds)
```

```python
chess = video_from_path('recordings/Easy1.mp4')
used_frames = cut_video(chess, average_foregrounds, 'Easy1', 10)
save_used_frames(used_frames, 'Easy1')
```
